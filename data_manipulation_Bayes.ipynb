```python
from sklearn.naive_bayes import CategoricalNB
import numpy as np

clf = CategoricalNB()
clf.fit(X, y)
```

```python
predicted=clf.predict(X_test)
predicted
```

### ewaluacja:

- macierz pomyłek
- metryki: dokładność,precyzja, czułość, f-miara

```python
clf.classes_
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, predicted) #labels=[">50K", "<=50K"])
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=clf.classes_)
disp.plot()
```

```python
predicted
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print(accuracy_score(y_test, predicted))
print(precision_score(y_test, predicted, pos_label=' >50K'))
print(recall_score(y_test, predicted, pos_label=' >50K'))
print(f1_score(y_test, predicted, pos_label=' >50K'))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predicted, labels=clf.classes_))
```
